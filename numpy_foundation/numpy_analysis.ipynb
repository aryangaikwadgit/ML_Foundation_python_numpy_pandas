```python
prices = house_df['SalePrice'].values
areas = house_df['GrLivArea'].values

# TASK A: Calculate key statistics
# 1. Mean, median, std of prices
print(f"Average sale price : {np.mean(prices).round(2)}")
print(f"Median of sale price : {np.median(prices)}")
print(f"Standard Deviation of sale price : {np.std(prices).round(1)}")

# 2. 25th, 75th percentiles
Q1 = np.percentile(prices,25)
Q3 = np.percentile(prices,75)
print(f"25-%ile of sale price : {Q1}")
print(f"75-%ile of sale price : {Q3}")

# 3. IQR (Q3 - Q1)
IQR = Q3-Q1
print(f"Interquartile Range of sale price : {IQR}")

# TASK B: Outlier detection 
# Method 1: IQR method (< Q1-1.5*IQR or > Q3+1.5*IQR)
upper = Q3 + (1.5 * IQR)
lower =  Q1 - (1.5 * IQR)
out1 = prices[(prices<lower) | (prices>upper)] 
out1
```
